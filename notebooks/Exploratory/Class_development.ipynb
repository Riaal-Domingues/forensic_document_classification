```python
import gensim
from gensim.models.phrases import Phrases, Phraser

# Build the bigram and trigram models
bigram = gensim.models.Phrases(new_mail.body_pos_tokens, min_count=1, threshold=1) # higher threshold fewer phrases.
trigram = gensim.models.Phrases(bigram[new_mail.body_pos_tokens], threshold=1)
# Faster way to get a sentence clubbed as a trigram/bigram
bigram_mod = gensim.models.phrases.Phraser(bigram)
trigram_mod = gensim.models.phrases.Phraser(trigram)

print(bigram_mod[new_mail.body_pos_tokens])

print(trigram_mod[new_mail.body_pos_tokens])
```
